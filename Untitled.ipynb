```python
#Подгружаю документ с выплаченными суммами

from docx import Document
from os import listdir

filename = [a for a in listdir() if a.endswith('.docx')][-1]

f = open(filename, 'rb')
document = Document(f)
f.close()
```

```python
#составляю списки по должностям

table = document.tables[1]

a = table.columns
teachers = a[2]
not_personal = []
director = []
engineer = []
assistents = []
dotsents = []
professors = []
st_prep = []
for row in teachers.cells:
    if 'не сотрудник спбгу' in row.text.lower():
        not_personal.append(row.text)
    if 'заместитель директора' in row.text.lower(): 
        director.append(row.text)
    if 'инженер' in row.text.lower():
        engineer.append(row.text)
    if 'ассистент' in row.text.lower():
        assistents.append(row.text)
    if 'старший' in row.text.lower():
        st_prep.append(row.text)
    if 'профессор' in row.text.lower():
        professors.append(row.text) 
    if 'доцент' in row.text.lower():
        dotsents.append(row.text)

director, engineer, assistents, dotsents, professors, st_prep
```

```python
# переводим всё в словарь: ключи это названия курсов\tкол-во модулей; значения -- список пар препод-заработок 
course_lectors = dict()

for row in table.rows[1:-1]:
    if not row.cells[1].text in course_lectors:
        course_lectors[row.cells[1].text + '\t' + row.cells[3].text] = [(row.cells[2].text, row.cells[4].text)]
    else:
        course_lectors[row.cells[1].text + '\t' + row.cells[3].text].append((row.cells[2].text, row.cells[4].text))
```

```python
for lector in course_lectors:
    course_name = lector.split('\t')[0]
    if 'физик' in course_name.lower():
        modules_number = 1
    modules_number = lector.split('\t')[1]
    try:
        int(modules_number)
    except:
        print(lector)
```
