```python
# Подготовка сведений о курсах и преподавателях

from docx import Document
from os import listdir

filename = 'gud.docx'
f = open(filename, 'rb')
document = Document(f)
f.close()

table = document.tables[1]

courses = dict()

for row in table.rows[1:]:
    if row.cells[3].text == '':
        row.cells[3].text = '1'
    if (row.cells[1].text, int(row.cells[3].text)) in courses:
        courses[(row.cells[1].text, int(row.cells[3].text))].append((row.cells[2].text, int(row.cells[4].text.replace(' ', ''))/1.271))
    else:
        courses[(row.cells[1].text, int(row.cells[3].text))] = [(row.cells[2].text, int(row.cells[4].text.replace(' ', ''))/1.271)]
```

```python
from docx import Document
from os import listdir
from docx.shared import Pt

COURSE_NAME_PARAGRAPH_INDEX = 8

# Для статистики соберем информацию о получившихся стоимостях. Они описывают прайс для каждого лектора
forums_costs = []
lections_costs = []
control_task_costs = []
structure_costs = []

for course in courses:
    for author in courses[course]:
        
        filename = 'template.docx'
        f = open(filename, 'rb')
        document = Document(f)
        f.close()
        
        # Общая стоимость ВСЕГО
        total_price = author[1]
        
        # Общие стоимости
        #forums_price = 2500 * course[1] # 2500 за модуль во всём курсе
        forums_price = round(2500 * course[1] / len(courses[course]), 2)
        lections_price = round(9 * (total_price - forums_price) / 11, 2)
        control_task_price = round((total_price - forums_price) / 11, 2)
        structure_price = round((total_price - forums_price) / 11, 2)
        
        # Часы
        #forums_hours = 1.75 * course[1]
        forums_hours = round(1.75 * course[1] / len(courses[course]), 2)
        lections_hours = 1.5 * 3 * course[1]
        control_task_hours = 0.5 * course[1]
        structure_hours = 0.5 * course[1]
        
        # Стоимости услуг
        forums_cost = round(forums_price / forums_hours, 2)
        lections_cost = round(lections_price / lections_hours, 2)
        control_task_cost = round(control_task_price / control_task_hours, 2)
        structure_cost = round(structure_price / structure_hours, 2)
        
        # Добавим получившиеся значения в общие списки, чтобы исследовать значения
        forums_costs.append(forums_cost)
        lections_costs.append(lections_cost)
        control_task_costs.append(control_task_cost)
        structure_costs.append(structure_cost)
        
        # Объём услуг и стоимость услуги ВСЕГО
        total_cost = forums_cost + lections_cost + control_task_cost + structure_cost
        total_hours = forums_hours + lections_hours + control_task_hours + structure_hours
        
        # Метаданные
        course_name = course[0]
        initials = author[0].split(' ')[1][0] + '. ' + author[0].split(' ')[2][0] + '. ' + author[0].split(' ')[0]
        surname = author[0].split(' ')[0]
        file_name = 'TZ_generated/ТЗ Контракт_44 ФЗ_' + surname + '.docx'
        
        # Запись полученных данных в таблицу с разбивкой по услугам
        pricelist = document.tables[0]
        
        pricelist.cell(1, 2).text = str(lections_hours)
        pricelist.cell(1, 4).text = str(lections_cost)
        pricelist.cell(1, 5).text = str(lections_price)
        
        pricelist.cell(2, 2).text = str(forums_hours)
        pricelist.cell(2, 4).text = str(forums_cost)
        pricelist.cell(2, 5).text = str(forums_price)
        
        pricelist.cell(3, 2).text = str(structure_hours)
        pricelist.cell(3, 4).text = str(structure_cost)
        pricelist.cell(3, 5).text = str(structure_price)
        
        pricelist.cell(4, 2).text = str(control_task_hours)
        pricelist.cell(4, 4).text = str(control_task_cost)
        pricelist.cell(4, 5).text = str(control_task_price)
        
        pricelist.cell(5, 2).text = str(total_hours)
        pricelist.cell(5, 4).text = str(total_cost)
        pricelist.cell(5, 5).text = str(round(total_price, 2))
        
        initials_table = document.tables[1]
        
        initials_table.cell(1, 1).text = initials_table.cell(1,1).text.replace('Автор_тут', initials)
        
        # Инициалы пишутся 11-ым шрифтом, чтобы документ не "плавал"
        initials_cell = initials_table.cell(1, 1)
        paragraphs = initials_cell.paragraphs
        for paragraph in paragraphs:
            for run in paragraph.runs:
                font = run.font
                font.size = Pt(11)        
        
        document.paragraphs[COURSE_NAME_PARAGRAPH_INDEX].text = document.paragraphs[COURSE_NAME_PARAGRAPH_INDEX].text.replace('название_курса', course_name) 
        
        document.save(file_name)
        
# Вывод отчёта
#print('Получившиеся стоимости')
#print(min(forums_costs), '-', max(forums_costs), '\t -- оплата ведения форумов;')
#print(min(lections_costs), '-', max(lections_costs), '\t -- оплата разработки лекций;')
#print(min(control_task_costs), '-', max(control_task_costs), '\t -- оплата разработки КЗ и ИЭ;')
#print(min(structure_costs), '-', max(structure_costs), '\t -- оплата разработки структуры.')
# Дорогие гитхабчане, эта информация для вас пусть останется закомментированной :)
```

```python
# Песочница

from docx import Document
from os import listdir
from docx.shared import Pt
from docx.enum.style import WD_STYLE_TYPE
from os import listdir

def col_sum_by_number(table, number):
    accumulator = 0
    for i in range(1, 5):
        accumulator += float(table.cell(i, number).text)
    return accumulator

for filename in listdir('TZ_generated'):
    filename = 'TZ_generated/' + filename
    # Открываем файл
    f = open(filename, 'rb')
    document = Document(f)
    f.close()
    
    # Выделяем таблицу с данными
    table = document.tables[0]
    
    # Сумма значений всех ячеек, кроме последней, в последнем столбце должна равняться значению в последней ячейке этого столбца
    if col_sum_by_number(table, 5) - float(table.cell(5, 5).text) > 1 or col_sum_by_number(table, 5) - float(table.cell(5, 5).text) < -1:
        print(filename.split('/')[1], 'is broken. Diffrence is ', sum([float(table.cell(i, 5).text) for i in range(1, 5)]) - float(table.cell(5, 5).text))
```
